```python
import pandas as pd
import numpy as np
df = pd.read_csv('bodyfat.csv')
df.tail()
```

```python
df['Weight'] = df['Weight'] * 0.45
df['Height'] = df['Height'] * 2.54
df
```

# Норамлизация 


```python
categorical_list = df.columns
categorical_list = categorical_list.to_list()


for i in categorical_list:
    df[i]=df[i]/np.max(df[i])

x=df[[
    'Density',
    # 'BodyFat',
    # 'Age',
    'Weight', 
    # 'Height',
    'Chest',
    #  'Abdomen',
    #  'Hip',
    #  'Thigh',
     'Knee',
    #  'Ankle',
     'Biceps',
     'Forearm',
     'Wrist'
]]
y= df.Neck
```

```python
df
```

# Модель


```python
from tensorflow import keras
from tensorflow.keras import layers
from sklearn.model_selection import KFold


input_shape=[7]

def baseline_model(): # лучшая 
    model =  keras.Sequential([
            layers.Dense(units=300, activation='relu', input_shape=input_shape),
            layers.Dense(units=230, activation='relu'),
            layers.Dense(units=100, activation='relu'),
            layers.Dense(units=80, activation='relu'),
            layers.Dense(units=50, activation='relu'),
            layers.Dense(units=20, activation='relu'),
             layers.Dense(units=10, activation='relu'),
        layers.Dense(units=1)])
    model.compile(
    optimizer='adam',
    loss="mean_squared_error",
)
    return model
```

```python
cv=KFold(n_splits=10, shuffle=True, random_state=0)
```

```python
from sklearn.model_selection import cross_val_score, cross_val_predict
from keras.wrappers.scikit_learn import KerasRegressor
```

```python
estimator = KerasRegressor(build_fn=baseline_model, epochs=550, batch_size=252, verbose=1)
results = cross_val_predict(estimator, x, y, cv=cv)
```

```python
from sklearn.metrics import r2_score #  epochs=550
r2_score(results, y)
```

Резултат нейронки 0.6501033972300192

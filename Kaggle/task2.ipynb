```python
def my_loss(y_true, y_pred):
    loss =   np.sum(np.logical_and((y_true==2), (y_true!=y_pred)))*5
    loss +=  np.sum((y_true!=2) & (y_true!=y_pred))
    loss/=len(y_true)
    return loss
```

```python
def predict(x, roof):
    ans=[]
    tmp= np.copy(x)
    for i in tmp:
        if i[2]>roof:
            ans.append(2)
        else:
            i[2]=0
            ans.append(np.argmax(i))
    return ans

```

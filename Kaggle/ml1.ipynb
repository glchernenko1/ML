```python
import pandas as pd
import seaborn as sns
import numpy as np
```

```python
df = pd.read_csv('dota2Train.csv')
df.describe(include='all')
```

```python
tmp = pd.value_counts(df['2'].values.ravel())
print(tmp)

tmp= np.array(tmp)
print((tmp[0]-np.sum(tmp[1:]))/len(df))
```

```python
from sklearn.model_selection import cross_val_score
from catboost import CatBoostClassifier
from sklearn.model_selection import StratifiedKFold
import lightgbm as lgb
import xgboost as xgb

cv = StratifiedKFold(n_splits=10, shuffle=True, random_state=0)
```

```python


model = CatBoostClassifier(iterations=1000,
                           task_type="GPU",
                           devices='0:1', max_depth=13)

prediction1 =  cross_val_score(model, df.loc[:, df.columns!='2'], df['2'], cv=cv)

print(prediction1)
print('CatBoostClassifier: {:6.4}'.format(np.mean(prediction1)))
```

```python
model21 = xgb.XGBClassifier(predictor = 'gpu_predictor') #max_delta_step = 2....15    !!! max_depth  = 2....15 

prediction2 =  cross_val_score(model21, df.loc[:, df.columns!='2'], df['2'], cv=cv)
print(prediction2)
print('XGBClassifier: {:6.4}'.format(np.mean(prediction2)))
```

```python
model22 = lgb.LGBMClassifier(verbosity=-1)
prediction3 =  cross_val_score(model22, df.loc[:, df.columns!='2'], df['2'], cv=cv)
print(prediction3)
print('LGBMClassifier: {:6.4}'.format(np.mean(prediction3)))
```
